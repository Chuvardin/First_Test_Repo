```

import pandas as pd     
import numpy as np      
import matplotlib.pyplot as plt 
```

```


url = "https://drive.google.com/uc?export=download&id=1R-bDTTCsLjzy9NZG3bkzzeNnFyBuupGK"


input_df = pd.read_csv(url, sep = ';', index_col = None) 

input_df.head()
```

```
# 3. Просмотр последних строк таблицы
input_df.tail()
```

```
# 4. Получение базовых статистических показателей
input_df.describe()
```

```
# 5. Можно сразу построить графики
axes = input_df.plot(marker=None, figsize=(20, 20), subplots = True)
```

```
# 6. Можно построить все вместе, заодно удалив индекс строк, так как его график не несет информации.

axes = input_df[input_df.columns[ 0 : 5 ]].plot(marker= '.', figsize=(20, 20), subplots = False)
```

```

forecast = input_df.iloc[: , 1].values
Station_A = input_df.iloc[: , 2].values
Station_B = input_df.iloc[: , 3].values
Station_c = input_df.iloc[: , 4].values
Station_D = input_df.iloc[: , 4].values


```

```
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width = 2000, plot_height = 600)


p.line(np.arange(len(forecast)), forecast, line_width = 5,  color = 'orange')

p.line(np.arange(len(Station_A)), Station_A, line_width = 2, color = 'red')


show(p)
data_frame = pd.DataFrame(forecast)
data_frame.to_csv('forecast_A.csv', index = False, header = ['forecast/St_A'])
data_frame.to_excel('forecast_A.xlsx', index = False, header = ['forecast/St_A'])
```
